```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.interpolation import map_coordinates
```

```python
def draw(r, title):
    r = np.array(r)
    rads = np.arange(0, 390, 30)
    rads = rads*np.pi/180

    A = np.array([rads, r], dtype=np.float32)
    A.shape

    # Interpolate values
    new_dims = []
    for original_length, new_length in zip(A.shape, (2,300)):
        new_dims.append(np.linspace(0, original_length-1, new_length))

    coords = np.meshgrid(*new_dims, indexing='ij')
    B = map_coordinates(A, coords)

    rads = B[0]
    r = B[1]
    
    plt.axes(projection = 'polar')
    plt.polar(rads, r, 'g.')
    plt.title(title)

        # display the Polar plot
    plt.show()
```

```python
draw([2, 2, 1, 1, 1, 1, 2, 1, 1, 1, 1, 1, 2], "YAGI Antenna 2-Element Without Reflector")
```

```python
draw([3, 2, 1, 0, 1, 1, 1, 1, 0, 0, 1, 1, 3], "YAGI Antenna 3-Element Both Reflector and Director")
```
